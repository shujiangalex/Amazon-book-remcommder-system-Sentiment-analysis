```python
df = pd.concat([df_b,df_c,df_h],ignore_index=True )
df.head()
```

```python
df['review rating'] =df['review rating'].astype(float)
```

```python
df = df[df['review rating'] != 3]
```

```python
df.loc[df['review rating'] > 3, 'Label'] = 1
df.loc[df['review rating'] < 3, 'Label'] = 0
```

```python
df['review_content'] =df['review_content'].astype(str)
```

```python
df['Label'] =df['Label'].astype(int)
df.head()
```

```python
df_class0 = df[df["Label"] == 0]
df_class1 = df[df["Label"] == 1]
```

```python
df_class1_sample = df_class1.sample(n = df_class0.shape[0], random_state = 203)
```

```python
df_new = pd.concat([df_class0, df_class1_sample], axis = 0, ignore_index = True)
```

```python
df_new.head()
```

# Dataset For Sentiment Analysis

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X = df_new["review_content"]
y = df_new["Label"]
```

```python
X.head()
```

# TF-IDF

```python
tfidf = TfidfVectorizer(min_df = 5, ngram_range= (1,1), max_df = 0.9)
X_tfidf= tfidf.fit_transform(X)
```

```python
X_tfidf.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size = 0.3, random_state = 203)
```

```python
X_train.shape
```

# XGBoost

```python
xgb = XGBClassifier(learning_rate=0.02, n_estimators=600,
                    objective='binary:logistic', random_state=203)
```

```python
xgb.fit(X_train, y_train)
pred_xgb = xgb.predict(X_test)
```

```python
roc_auc_score(pred_xgb, y_test)
```

```python
accuracy_score(pred_xgb, y_test)
```

```python
confusion_matrix(pred_xgb, y_test)
```

```python
plot_roc_curve(xgb, X_test, y_test)
```
